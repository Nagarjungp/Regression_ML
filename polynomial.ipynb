```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
target = pd.DataFrame(boston.target, columns=['TARGET'] )
```

```python
data=pd.concat([features, target], axis=1)
```

```python
data.head(30)
```

```python
sep = 0.8*len(data)
sep = int(sep)

x_train = np.array(data['RM'] [:sep])
y_train = np.array(data['TARGET'] [:sep])

x_test=np.array(data['RM'][sep:])
y_test=np.array(data['TARGET'][sep:])

x_train=x_train/x_train.mean()
y_train=y_train/y_train.mean()

x_test=x_test/x_test.mean()
y_test=y_test/y_test.mean()
```

```python
plt.plot(x_train,y_train,'b.')
```

```python
def hypothesis(a,b,c,x):
    return (a*x*x+b*x+c)
```

```python
def error(a,b,c,x,y):
    e=0
    m=len(y)
    for i in range(m):
        e +=np.power(hypothesis(a,b,c,x[i])-y[i],2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,b,c,x,y,learning_rate):
    grad_a=0
    grad_b=0
    grad_c=0
    m=len(y)
    for i in range(m):
        grad_a += 2/m *((hypothesis(a,b,c,x[i])-y[i]) * x[i]*x[i])
        grad_b += 2/m *(hypothesis(a,b,c,x[i])-y[i]) * x[i]
        grad_c += 2/m *(hypothesis(a,b,c,x[i])-y[i])
        
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    c=c-(grad_c*learning_rate)
    
    return a,b,c
```

```python
def descend(initial_a,initial_b,initial_c,x,y,learning_rate,iterations):
    a=initial_a
    b=initial_b
    c=initial_c
    for i in range(iterations):
        e = error(a,b,c,x,y)
        a,b,c = step_gradient(a,b,c,x,y,learning_rate)
        if i % 1000 == 0:
            print(f"Error:{e}, a:{a}, b:{b}, c:{c}")
        
                  
    return a,b,c
```

```python
init_a=0
init_b=0
init_c=0
iters=10000
learning_rate=0.01
```

```python
final_a,final_b,final_c=descend(init_a,init_b,init_c,x_train,y_train,learning_rate,iters)
```

```python
plt.plot(x_train,y_train,'b.', x_train, hypothesis(init_a,init_b,init_c,x_train), 'g')
plt.plot(x_train,hypothesis(final_a,final_b,final_c,x_train), 'r.')
```
