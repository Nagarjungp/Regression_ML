```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston 
```

```python
boston = load_boston()
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
target = pd.DataFrame(boston.target, columns=['TARGET'] )
```

```python
data=pd.concat([features, target], axis=1)
```

```python
sep = 0.8*len(data)
sep = int(sep)

x_train = np.array(data['LSTAT'] [:sep])
y_train = np.array(data['TARGET'] [:sep])

x_test=np.array(data['LSTAT'][sep:])
y_test=np.array(data['TARGET'][sep:])

x_train=x_train/x_train.mean()
y_train=y_train/y_train.mean()

x_test=x_test/x_test.mean()
y_test=y_test/y_test.mean()
```

```python
plt.plot(x_train,y_train,'r.')
plt.xlabel('l stat')
plt.ylabel("price")
```

```python
def hypothesis(a,x,b):
    return b*np.exp(-a*x)
```

```python
def error(a,x,b,y):
    e=0
    m=len(y)
    for i in range(m):
        e +=np.power(hypothesis(a,x[i],b)-y[i],2)
        
    return (1/(2*m))*e
```

```python
def step_gradient(a,x,b,y,learning_rate):
    grad_a = 0
    grad_b = 0
    m = len(x)
    for i in range(m):
        grad_a +=-2/m * (hypothesis(a,x[i],b) -y[i]) * x[i]
        grad_b +=2/m * (hypothesis(a,x[i],b) -y[i]) 
        
    a=a-(grad_a * learning_rate)
    b=b-(grad_b * learning_rate)
    
    return a,b
```

```python
def descend(initial_a, initial_b, x, y, learning_rate):
    a=initial_a
    b=initial_b
    i=0
    
    while True:
        e1=error(a,x,b,y)
        a,b = step_gradient(a,x,b,y,learning_rate)
        print(f"error: a:{e1}, a:{a},b:{b}")
        e2=error(a,x,b,y)
        
       
        
            
       
        if e2>e1:
            break
        
    return a,b 
```

```python
init_a=0
init_b=1

learning_rate=0.1
```

```python
final_a,final_b=descend(init_a,init_b, x_train,y_train,learning_rate)
```

```python
plt.plot(x_train, y_train, 'r.', x_train, hypothesis(init_a, x_train, init_b), 'b')
plt.plot(x_train, hypothesis(final_a, x_train, final_b), 'g.')
```
