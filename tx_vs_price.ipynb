```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
target = pd.DataFrame(boston.target, columns=['TARGET'] )
```

```python
data=pd.concat([features, target], axis=1)
```

```python
sep=0.8*len(data)
sep=int(sep)

x_train=np.array(data['TAX'] [:sep])
y_train=np.array(data['TARGET'] [:sep])

x_test=np.array(data['TAX'] [sep:])
y_test=np.array(data['TARGET'] [sep:])

x_train=x_train/x_train.mean()
y_train=y_train/y_train.mean()

x_test=x_test/x_test.mean()
y_test=y_test/y_test.mean()
```

```python

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.plot3D(x_train,y_train,'g.')
```

```python
def hypothesis(a,b,x):
    return a*x+b
```

```python
def error(a,b,x,y):
    e=0
    m=len(y)
    for i in range(m):
        e += np.power(hypothesis(a,b,x[i])-y[i],2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,b,x,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(y)
    for i in range(m):
        grad_a += 2/m * (hypothesis(a,b,x[i])-y[i]) * x[i]
        grad_b += 2/m * (hypothesis(a,b,x[i])-y[i]) 
        
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    
    return a,b
```

```python
def descend(initial_a,initial_b,x,y,learning_rate,iterations):
    a=initial_a
    b=initial_b
    for i in range(iterations):
        e = error(a,b,x,y)
        a,b=step_gradient(a,b,x,y,learning_rate)
        if i%1000 == 0:
            print(f"Error:{e}, a:{a}, b:{b}")
            
    return a,b
            
```

```python
a=0
b=1
learning_rate=0.01
iters=10000
```

```python
final_a,final_b=descend(a,b,x_train,y_train,learning_rate,iters)
```
